```python
import pandas as pd

dic = {"이름": ["청수", "영희", "민철", "경수", "동수", "소연", "민수"],
       "학번": [1000, 1001, 1002, 1003, 1004, 1005, 1006],
       "지역": ["서울", "대전", "대구", "부산", "서울", "제주", "서울"],
       "국어": [84, float("NaN"), 81, 80, 97, float("NaN"), 99],
       "영어": [81, 91, 97, 99, 96, float("NaN"), 85],
       "수학": [87, float("NaN"), 86, 76, float("NaN"), 90, 95],
       "과학": [74, 83, 75, 71, 98, float("NaN"), 80]}
df = pd.DataFrame(dic)

df = df.fillna(50)

df["국어"] = df["국어"].fillna(df["국어"].mean())
df["영어"] = df["영어"].fillna(df["영어"].max())
df["수학"] = df["수학"].fillna(df["수학"].min())
df["과학"] = df["과학"].fillna(0)
```
