```python
import numpy as np
import random
list_1 = []

for i in range(20):
    list_1.append(random.randint(1,100))

rarr = np.array(list_1)
arr2 = rarr.reshape(5, -1)

print(arr2)

print("=========== axis = 0 행기준 위아래 ===========")
ar1 = np.sort(arr2, axis = 0)
print(ar1)

print("=========== axis = 1 열기준 좌우 ===========")
ar2 = np.sort(arr2, axis = 1)
print(ar2)
```
