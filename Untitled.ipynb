```python
import numpy as np
x = [1,2,3,4,5,6]

arr1 = np.array(x, int)

print(arr1, type(arr1))

print(arr1.dtype)

print("shape = {}, ndim = {}, size = {}".format(arr1.shape, arr1.ndim, arr1.size))
```

```python
import numpy as np
x1 = [[1,5],[3,7],[13,2],[7,10]]

arr2 = np.array(x1, float)

print(arr2, type(arr2))

print(arr2.dtype)

print("shape = {}, ndim = {}, size = {}".format(arr2.shape, arr2.ndim, arr2.size))
```

```python
import numpy as np

list_1 = [18, 71, 45, 33, 65, 47, 9 ,24, 36]

sum = 0

for i in list_1:
    sum = sum + i

average = sum / len(list_1)

print("sum = {}".format(sum))
print("average = {}".format(average))

print("\n====== 절 취 선 ======\n")

arr1 = np.array(list_1)

print("sum = ", np.sum(arr1))
print("average = ", np.average(arr1))
print("median = ", np.median(arr1))
print("var = ", np.var(arr1))
print("std = ", np.std(arr1))
print("max = ", np.max(arr1))
print("min = ", np.min(arr1))
```
