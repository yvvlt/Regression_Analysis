```python
import numpy as np

scores = [88, 92, 75, 100, 89, 95, 73, 68, 85, 90, 78, 82, 96, 87, 91]

arr = np.array(scores) # numpy 배열 변환

mean_score = np.mean(arr) # 평균
max_score = np.max(arr) # 최댓값
min_score = np.min(arr) # 최솟값
median_score = np.median(arr) # 중앙값

print(arr[arr >= 80]) # 80점 이상 점수출력
count_below_70 = np.sum(arr <= 70) # 70점 이하
sorted_scores = np.sort(arr) # 오름차순

# 5점 추가 (100점 초과x)
added_scores = np.minimum(arr + 5, 100)  
```
