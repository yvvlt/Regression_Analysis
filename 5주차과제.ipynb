```python
# 실습예제 1
import numpy as np

scores = [88, 92, 75, 100, 89, 95, 73, 68, 85, 90, 78, 82, 96, 87, 91]

arr = np.array(scores) # numpy 배열 변환

mean_score = np.mean(arr) # 평균
max_score = np.max(arr) # 최댓값
min_score = np.min(arr) # 최솟값
median_score = np.median(arr) # 중앙값

print(arr[arr >= 80]) # 80점 이상 점수출력
count_below_70 = np.sum(arr <= 70) # 70점 이하
sorted_scores = np.sort(arr) # 오름차순

# 5점 추가 (100점 초과x)
added_scores = np.minimum(arr + 5, 100)  
```

```python
# 실습예제 2
import numpy as np

class_A = np.array([78,75,90,88,76,95,89,84,91,87,80,93,72,99,83,77,94,86,81,79])
class_B = np.array([82,79,88,92,75,89,90,80,86,85,78,96,74,100,82,76,97,84,83,81])

# 평균
mean_A = np.mean(class_A)
mean_B = np.mean(class_B)

# 중앙값, 표준편차
median_A = np.median(class_A)
median_B = np.median(class_B)
std_A = np.std(class_A, ddof=0)
std_B = np.std(class_B, ddof=0)

# 최상위 5명 점수
top5_A = np.sort(class_A)[-5:]
top5_B = np.sort(class_B)[-5:]

mean_top5_A = np.mean(top5_A)
mean_top5_B = np.mean(top5_B)

# 80점 이상 비율
ratio_80_A = np.sum(class_A >= 80) / len(class_A)
ratio_80_B = np.sum(class_B >= 80) / len(class_B)
```
